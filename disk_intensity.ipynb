```python
import matplotlib.pyplot as plt
import numpy as np

fig, axs = plt.subplots(4, 3, figsize=(12, 16))
plt.subplots_adjust(left=0.125, bottom=0.1, right=0.9, top=0.9, wspace=0.2, hspace=0.2)

sz = []
with open('disk_intensity.in', 'r') as f:
    t = int(f.readline().strip())
    assert t == 12
    for i in range(t):
        sz.append(int(f.readline().strip()))

with open('disk_intensity.out', 'r') as f:
    for i in range(12):
        trials = int(f.readline().strip())
        mat = [[-float(val) for val in f.readline().strip().split()] for _ in range(2 * sz[i] + 1)]
        f.readline()

        axs[i // 3][i % 3].imshow(mat, cmap='PuOr')
plt.show();
```
